```python
import seaborn
import pandas as pd
import matplotlib.pyplot as plt
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
dados = pd.read_csv("SINASC_RO_2019.csv")
seaborn.boxplot(y='IDADEMAE', data=dados)
```

####  2) Faça um boxplot do peso do bebê

```python
seaborn.boxplot(y='PESO', data=dados)
```

####  3) Faça um histograma da idade da mãe

```python
seaborn.displot(data = dados 
            ,alpha = .20
            ,x = 'IDADEMAE'
            ,bins = 40
            ,element = 'step'
            ,kde = True
)
```

#### 4) Faça um histograma do peso do bebê

```python
seaborn.displot(data = dados 
            ,alpha = .20
            ,x = 'PESO'
            ,bins = 50
            ,element = 'step'
            ,kde = True
)
```
